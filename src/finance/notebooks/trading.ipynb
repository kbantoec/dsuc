```python
tsla: pl.DataFrame = pl.read_csv("../data/raw/TSLA.csv", parse_dates=True)
display(tsla.head())
```

```python
tsla.dtypes
```

# Computing returns

```python
tsla = tsla.with_column(pl.col("Close").pct_change(n=1).map(lambda r: r * 100).alias("daily_return"))
display(tsla.head())
```

```python
fig = px.histogram(x=tsla.select("daily_return").to_series(),
                   nbins=100,
                   labels={"x": "Daily Tesla returns (in %)"},
                   histnorm="probability density")
# fig.update_layout(
#     yaxis_title_text="Probability density (the sum of all bar areas equals 1)",
#     bargap=0.1
# )
fig.show()
```
